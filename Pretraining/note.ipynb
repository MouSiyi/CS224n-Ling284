## Word structure and subwords
Some assumptions we've made about a language's vocabulary:
- A fixed vocab of words built from the training set. All novel words are mapped to a single UNK at test time. 

So basically we want a more robust vocab.

We can make use of morphology to do subword modeling.
- The dominant modern paradigm is to learn a vocab of subword tokens
- At training and testing time, each word is split into a sequence of known subwords.

## Pretraining
### Motivation
"...the complete meaning of a word is always contextual"

With word2vec only, it only builds embeddings "statically" without context. So word embeddings are called to be pretrained.      
With encoders, we can get contextual representation of words that are not pretrained.

All/almost all parameters in nn are intialized via pretraining.

### Pretraining through LM
Language modeling task: model the conditional probability
$$p_{\theta}(w_t|w_{1:t-1})$$

Pretraining on language modeling 
- Train a nn to perform language modeling on a large amount of text
- Save the netword parameters   
 
Use pretraining parameters as parameter initialization for fine-tuning


- The underlying point here is that the starting point matters(but who knows why?)

### Encoders
Encoders get bidirectional context, so the objective for language modeling does not make sense here. We cannot do language modeling.

Objective idea: [MASK] some fraction of tokens, and predict them.
Let $\tilde x$ be the masked version of $x$, then we're actually learning $p_{\theta}(x|\tilde x)$.
