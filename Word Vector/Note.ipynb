## Representing Meaning
### Denotational Semantics
In denotational semantics, a meaning is a pairing(mapping) of 
$$signified(idea\ or\ thing)\rightarrow signifier(symbol)$$

Common NLP solution to construct meaning in computer is by using thesaurus containing synonyms(同义词) and hypernyms(上位词 is a).
Cons:     
- Lacking nuance
- Subjective and manual
- Unsuitable for further quantitizing like computing word similarity.

In traditional NLP, words are regarded as discrete symbols(localized representation).    
We represent them by one-hot vectors.    
Cons:      
- One-hot vectors are orthogonal
- No natural notion of similarity among the vectors

Solution: **learn** to encode similarity in the vectors themselves

### Distributional Semantics
Idea: A words's meaning is given by the words that frequently appear close-by.    
Context: the set of words appear near by the word $w$.    
We can use the many contexts of $w$ to construct the meaning representation.


### Word Vectors/Embedding

### Word2vec
A framework for learning word vectors.

#### Idea
We have a large corpus of text.    
Then we choose a fixed vocabulary(large but truncated), create a vector for each word.    
We learn the vectors from the text by maximizing the probability of  *o*(outside) given *c*(center) or vice versa, going through each position in the text.

#### Objective function
$$Likelihood = L(\theta) = \prod_{t=1}^T\prod_{-m\le j\le m,j\ne 0}P(w_{t+j}|w_t;\theta)$$
$$Loss = l(\theta) = -\frac1T\sum_{t=1}^T\sum_{-m\le j\le m,j\ne 0}logP(w_{t+j}|w_t;\theta)$$

Further we need to calculate $P(w_{t+j}|w_t;\theta)$.    
We give each word $w$ two vectors:
- $v_{w}$ when $w$ is a center word.
- $u_{w}$ when it is a outside word.
$$ P(o|c) = \frac{exp(u_o^Tv_c)}{\sum_{w\in V}exp(u_w^Tv_c)}$$
It is actually a softmax function mapping $R^n$ to $(0, 1)$, returning a probability distribution.     
In this design, we compare similarity by doing doe product and map the "scores" into a distribution.    

#### Optimization
$\theta$ represents all the model parameters in one long vector.    
- $V$ words, 
- For each word, 2 vectors
- For each vector, $d$ dimensions
